```julia
using SymPy
using Interact
using PyPlot
using PyCall
using Markdown
using ODE
@pyimport matplotlib.animation as anim 
```

```julia
function x(q0)
    q0[3]-q0[1]
end

function y(q0)
    q0[4]-q0[2]
end  

function r(q0)
    sqrt(x(q0)^2+y(q0)^2)
end

function a(t,q0,v0,m1,m2,G)
    (G/(r(q0)^3)).*[m2*x(q0),m2*y(q0),-m1*x(q0),-m1*y(q0)]
end
```

```julia
function Euler(q0,v0,ti,h,tf,f,m1=1,m2=1,G=1)
    vs=Vector{Float64}[]
    ts = Float64[]
    qs=Vector{Float64}[]
    for t in ti:h:tf
        x_nueva=q0+v0*h
        v_nueva=v0+a(t,q0,v0,m1,m2,G)*h   
        q0=x_nueva
        v0=v_nueva
        push!(qs,[q0])
        push!(vs,[v0])
        push!(ts,t)
    end
    ts,hcat(qs...)',hcat(vs...)'
end
      
```

```julia
TS,QS,VS=Euler([0,0,1,0],[0,0,0,0.7],0,0.1,20,a,1,1e-5 ,1);
```

```julia
X1,Y1,X2,Y2=QS[:,1],QS[:,2],QS[:,3],QS[:,4];
```

```julia
RS=sqrt((X2-X1).^2+(Y2-Y1).^2);
```

```julia
plot(TS,RS)
xlabel(L"$tiempo$")
ylabel(L"$radio$")

    
```

```julia
fig2 = figure()
ax = plt.axes(xlim=(-2, 2), ylim=(-2, 2))

global line1 = ax[:plot]([], [], "bo", lw=2)[1]

function init()
    global line1

    line1[:set_data]([], [])

    return ([line1],None)
end

X1,Y1,X2,Y2=QS[:,1],QS[:,2],QS[:,3],QS[:,4]


function animate2(i)
    global line1
    line1[:set_data]([X1[i+1],X2[i+1]],[Y1[i+1],Y2[i+1]])
    return ([line1],None)
end;
```

```julia
myanim2 = anim.FuncAnimation(fig2, animate2, init_func=init, frames=100, interval=20, blit=true);
myanim2[:save]("planetas.gif", writer="imagemagick", fps=30);
```
