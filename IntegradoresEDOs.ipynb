### $i)$ Define funciones para hacer *un solo paso* de los métodos de Euler, punto medio, y RK4. Aceptan la función que integrar, el valor actual de la variable independiente, y el tamaño de paso, y regresan el nuevo valor de la variable independiente.

```julia
function Euler(fx,xi,dt)   
    x_nueva=xi+fx(xi)*dt 
    x_nueva
end;
```

```julia
function PM(fx,xi,dt)
    x_nueva=xi+fx(xi+((dt/2).*fx(xi)))*dt
    x_nueva
end;
```

```julia
function RK4(fx,xi,dt)
    k1(xi)=fx(xi)
    k2(xi,dt)=fx(xi+((k1(xi)*dt)/2))
    k3(xi,dt)=fx(xi+k2(xi,dt)*dt/2)
    k4(xi,dt)=fx(xi+k3(xi,dt)*dt)
    x_nueva=xi+((dt/6)*((k1(xi))+2*k2(xi,dt)+2*k3(xi,dt)+k4(xi,dt)))
    x_nueva
    
end;
```

### $ii)$ Define una función `integrar` que acepta un método de integración que usar, una función, el rango de tiempos y el tamaño del paso.

```julia
function integrar(metodo,fx,ti,dt,tf,xi)
    xs=Vector{Float64}[]
    ts=Vector{Float64}[]
    while ti<=tf
        x_nueva=metodo(fx,xi,dt)
        xi=x_nueva
        ti=ti+dt
        push!(xs,[xi])
        push!(ts,[ti])
    end
    hcat(xs...)',ts
end;
```

### $iii)$ Compara el funcionamiento de estos 3 métodos para el oscilador armónico.

```julia
function oscilador(v)
    θ = v[1]
    ω = v[2]
    [ω, -θ]
end
```

```julia
EOS=integrar(Euler,oscilador,0,0.1,33,[0,1]);
PMOS=integrar(PM,oscilador,0,0.1,33,[0,1]);
RK4OS=integrar(RK4,oscilador,0,0.1,33,[0,1]);
```

```julia
plot(EOS[2],EOS[1][:,1],label="euler", "-g")
plot(PMOS[2],PMOS[1][:,1],label="Punto Medio", "-r")
plot(RK4OS[2],RK4OS[1][:,1],label="RK4", "-b")
legend()
ylabel(L"$Osciones$")
xlabel(L"$t$")
```
