```julia
function f(α)
    xs=Float64[]
    ys=Float64[]
    ts=Float64[]
    for t in 0:π/360:π
        x=cos(6*t*π) 
        y=cos(4*π*t+π/2)
        push!(xs,x)
        push!(ys,y)
        push!(ts,t)
end
    ts,xs,ys
end
```
