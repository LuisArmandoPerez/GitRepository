```julia
using PyCall
using PyPlot
using Markdown


@pyimport matplotlib.animation as anim 
```

```julia
fig = figure()
ax = plt.axes(xlim=(-200, 200), ylim=(-200, 200))
global line1 = ax[:plot]([], [], "bo", lw=2)[1]
#global line2 = ax[:plot]([], [], "yo", lw=2)[1]

function init()
    global line1
 #   global line2

    line1[:set_data]([], [])
  #   line2[:set_data]([], [])

        return ([line1],None)
   
end

t=[0:pi/48:4*pi];
x=cos(t*3)+sin(4.15*t);
y=-sin(t*3)+cos(4.15*t);

function animate(i)
    global line1
    line1[:set_data]([x[i+1]],[y[i+1]])
      #line2[:set_data](0,0)
    return ([line1],None)

end
```

```julia
myanim = anim.FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=true);
myanim[:save]("mercurio.gif", writer="imagemagick", fps=30);
```
